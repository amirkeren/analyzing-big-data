### Loading the dataset

```python
rdd = sc.textFile('AtchisonUV_20150801_to_20151119.csv')
header = rdd.first()
print header
rdd = rdd.filter(lambda line: line != header)
```

### Transforming the data to RDD pairs

```python
pairs = rdd.map(lambda x: (x.split(",")[0], ','.join(x.split(",")[1:])))
pairs.count()
```

### Counting distinct values

```python
print 'Benzene', pairs.map(lambda x: (x[1].split(',')[0])).distinct().count()
print 'Xylene', pairs.map(lambda x: (x[1].split(',')[4])).distinct().count()
print 'Wind Direction', pairs.map(lambda x: (x[1].split(',')[6])).distinct().count()
print 'Wind Speed', pairs.map(lambda x: (x[1].split(',')[7])).distinct().count()
print 'Wind Origin', pairs.map(lambda x: (x[1].split(',')[8])).distinct().count()
```

### Distributions
